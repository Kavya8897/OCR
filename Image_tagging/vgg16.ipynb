```python
#Importing requirements

import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
# Downloading the vgg16 model

model = VGG16()
```

```python
model.summary()
```

```python
import os
```

```python
# reading all the image dataset and tagging 

img_size = 224
for file in os.listdir("data"):
    print(file)
    full_path ="C:/Users/Family/Documents/tasks/Image tagging ocr/image_tagging/data/" + file
    
    image = load_img(full_path,target_size = (img_size,img_size))
    image = img_to_array(image)
    image = image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))
    image = preprocess_input(image)
    y_predict = model.predict(image)
    label = decode_predictions(y_predict, top = 5)
    print(label)
    print()
```
